```python
# load text of macbeth from s3
#!wget https://dsr-notebooks.s3.amazonaws.com/macbeth.txt
shakespeare = sc.textFile("macbeth.txt")
```

```python
lines = shakespeare.map(lambda line: line.lower())
```

```python
lines.cache()
```

```python
macbeth = lines.filter(lambda l: l.find("macbeth") >= 0)
import sys
print(macbeth)
```

```python
macbeth.count()
```

```python
macbeth.take(5)
```

```python
macbeth_and_macduff = macbeth.filter(lambda l: l.find("macduff") >= 0)
```

```python
print(macbeth_and_macduff)
macbeth_and_macduff.count()
```

```python
import re
words = lines.flatMap(lambda l: re.split(r'\W+', l))  ## flatMap is similar to mp, it returns a new RDD by applying 
                                                      ## a function to each element of the RDD, but output is flattened
```

```python
groupedWords = words.map(lambda w: (w, 1))   ### Assign each word to an integer  1
unsorted = groupedWords.reduceByKey(lambda a, b: a + b)
unsorted.collect()
```

```python
unsorted.map(lambda t: (t[1], t[0])).sortByKey().collect() ## sortByKey() sorts by ascending order
```

```python
unsorted.map(lambda t: (t[1], t[0])).sortByKey(False).collect() ## sortByKey(False) sorts by descending order
```
